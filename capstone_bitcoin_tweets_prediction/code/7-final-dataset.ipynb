```python
import pandas as pd
import numpy as np
```

```python
btc = pd.read_csv('../data/1-bitcoin_data.csv')
tw_d = pd.read_csv('../data/6-final-tweet-day.csv')
tw_h = pd.read_csv('../data/6-final-tweet-hour.csv')
data = [btc,tw_d,tw_h]
for item in data:
    item.drop(columns='Unnamed: 0',inplace=True)
    item['timestamp'] = pd.to_datetime(item['timestamp'])
```

```python
tw_d['timestamp'] = pd.to_datetime(tw_d['timestamp']).dt.tz_localize(None)
tw_h['timestamp'] = pd.to_datetime(tw_h['timestamp']).dt.tz_localize(None)
```

```python
#concatinate btc prices and tweets on timestamp
df_d = tw_d.merge(btc, how="left", on="timestamp")
df_h = tw_h.merge(btc, how="left", on="timestamp")
df_d.dropna(axis=0, subset=['Close'],inplace=True)
df_h.dropna(axis=0, subset=['Close'],inplace=True)
df_d.to_csv('../data/7-final-data-day.csv')
df_h.to_csv('../data/7-final-data-hour.csv')
df_h.head()
```
